To ensure that the test set is representative of true data we do **stratified sampling** on the variabile *median_income*. We need to divide the variable in various _strata_, *i.e.* homogeneous subpopulations. Then we sample each stratum independently.

```python
df.hist(column='median_income', bins=200);
```

First, need to convert median income to categorical. _ceil_ rounds to larger integer. Income is in scale 0 - 15 so I'll have at the most 15 categories. Rescaling by 1.5 reduces number of categories

```python
df['median_income_cat'] = np.ceil((df['median_income'] / 1.5))
```

```python
df['median_income_cat'].hist(bins=20)
```

Need to have enough samples in each category so group together everything larger than 6.

```python
df['median_income_cat'].where(df['median_income_cat'] < 6, 6, inplace=True)
df['median_income_cat'].hist(bins=20)
```

OK I got overall 6 strata, no need to do stratified sampling.

```python
# Init splitter
splitter = sklearn.model_selection.StratifiedShuffleSplit(n_splits=1, random_state=42, test_size=0.1)
```

Loop is over splits. Could have used

`train_idxs, test_idxs = list(splitter.split(df, df['median_income_cat']))`

```python
for train_idxs, test_idxs in splitter.split(df, df['median_income_cat']):
    strat_train_set =  df.iloc[train_idxs] 
    strat_test_set =  df.iloc[test_idxs]
    print(".")
```

```python
strat_train_set.hist(bins=100, figsize=(10,10));
```

```python
strat_test_set.hist(bins=100, figsize=(10,10));
```

Data percentages in median income categories. To be compared to test set

```python
vals = (df['median_income_cat'].value_counts() / len(df))
vals
```

```python
test_vals = strat_test_set['median_income_cat'].value_counts() / len(strat_test_set)
test_vals
```

```python
vals - test_vals
```

Now drop the income category from data set (`axis` must be 1 to mean drop column)

```python
for dataset in (strat_train_set, strat_test_set):
    dataset.drop(['median_income_cat'], inplace=True, axis=1)
```
