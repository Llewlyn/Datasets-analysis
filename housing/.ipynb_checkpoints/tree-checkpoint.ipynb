```python
%clear
%store -z
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn as sk
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
plt.style.use('dark_background')
```

Load dataset

```python
data = np.load('data_np.npz')
# data.keys()
X_trn = data['arr_0']
X_tst = data['arr_1']
y_trn = data['arr_2']
y_tst = data['arr_3']
ix = data['arr_4'].item()
columns = [s for s in ix.keys()]
# np.load('columns.npy').item()
```

```python
from sklearn import tree


tree_reg = tree.DecisionTreeRegressor(max_leaf_nodes=5)
tree_reg.fit(X_trn, y_trn)
```

```python
tree_reg.tree_.node_count  # no. nodes in tree
tree_reg.tree_.feature  # feature used in each node (-2 = leaf)
tree_reg.tree_.threshold  # threshold used in each node (-2 = leaf)
tree_reg.tree_.impurity
tree_reg.tree_.compute_feature_importances()
```

```python
import graphviz


dot_data = tree.export_graphviz(
    tree_reg, 
    feature_names=columns,
    out_file=None,
    filled=True,
    proportion=True,
    impurity=True
)
graph = graphviz.Source(dot_data)
graph
```
